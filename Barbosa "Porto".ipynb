```python
from itertools import permutations, product

def generate_permutations(prepositions, main_words):
    # Generate all permutations of main words
    for main_perm in permutations(main_words):
        # Generate the Cartesian product of prepositions (with repetition) for 4 prepositions
        for prep_comb in product(prepositions, repeat=4):
            line = []
            for i in range(len(main_words)):
                line.append(prep_comb[i])
                line.append(main_perm[i])
            print(' '.join(line))

prepositions = ['o', 'da', 'no', 'na']
main_words = ['granita', 'saudade', 'pedra', 'historia']

generate_permutations(prepositions, main_words)
```

```python
# instafriendly version

from itertools import permutations, product
from PIL import Image, ImageDraw, ImageFont
import random

def generate_permutations(prepositions, main_words, background_color, font_path, font_size, font_color, max_lines):
    lines = []
    # Generate all permutations of main words
    for main_perm in permutations(main_words):
        # Generate the Cartesian product of prepositions (with repetition) for 4 prepositions
        for prep_comb in product(prepositions, repeat=4):
            line = ' '.join([f"{prep_comb[i]} {main_perm[i]}" for i in range(len(main_words))])
            lines.append(line)

    # Shuffle the lines and select a random subset
    random.shuffle(lines)
    lines = lines[:max_lines]

    # Create an image with the specified background color
    image_width = 1080  # Instagram story width
    image_height = 1920 # Instagram story height
    image = Image.new('RGB', (image_width, image_height), background_color)
    draw = ImageDraw.Draw(image)

    # Load the specified font and size
    font = ImageFont.truetype(font_path, font_size)

    # Define the margin and initial vertical position
    margin = 100
    y_text = margin

    # Write the lines to the image, omitting any that are too long
    for line in lines:
        text_size = draw.textsize(line, font=font)
        if text_size[0] + 2 * margin > image_width or y_text + text_size[1] > image_height - margin:
            continue # Skip this line if it's too long or if it exceeds the height
        draw.text((margin, y_text), line, font=font, fill=font_color)
        y_text += text_size[1] + 10 # 10 is line spacing

    # Save the image to a file
    image.save('porto.png')

prepositions = ['o', 'da', 'no', 'na']
main_words = ['granita', 'saudade', 'pedra', 'historia']
background_color = 'black'
font_path = '/Users/Sim/Library/Fonts/vt323/VT323-Regular.ttf' # Change this to the path of any TTF font file you prefer
font_size = 50
font_color = 'white'
max_lines = 50 # Maximum number of random lines to consider

generate_permutations(prepositions, main_words, background_color, font_path, font_size, font_color, max_lines)
```
